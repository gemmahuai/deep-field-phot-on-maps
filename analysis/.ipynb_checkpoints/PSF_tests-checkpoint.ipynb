__Problem...__

    Observe much broader PSFs after averaging azimuthally due to misaligned PSF center during rotation. 
    
    Specifically, the input PSFs are placed slightly off from the image center, as apparent from the brightest central spot of an input PSF is not coincident with the center of the 2D square array. In calculating azimuthally averaged PSFs, each asymmetric PSF is rotated multiple times covering 2pi range by fixing the rotation center at the image center. Note that by not fixing the rotation axis at the true PSF center, we're effectively broadening the resultant PSF. 

__Outcome?__

    This may or may not be problematic, depending on how galaxies are convolved with PSFs making exposures in the first place. If not corrected for each exposure, an azimuthally averaged (symmetrical) PSF would result in a source being placed at a false position. But in the deep field region, hundreds of exposures with different orientations are averaged to a mosaic, with the same gist of fixing the rotation axis described earlier. Thus, as long as the way we average PSFs is consistent with PSF convolution when generating exposures (convcube.py), we are good to go. 

## Interpolation

```python
from scipy.interpolate import interpn


### interpolate the SPHEREx PSFs
def interp_images_1d(imgs , Z1 , Z1_interp):
    ''' 
    Interpolates an array of images in 1 dimension
    
    INPUT:
    - imgs: an array of NxN images of shape (z , N , N), where z is the 3rd dimension (= number of images).
    - Z1: the axis of 3rd dimension (the simplest is np.arange(number_of_images))
    - Z1_interp: where to interpolate in 3rd dimension (relative to Z1). 
    
    OUTPUT:
    The interpolated NxN image.
    '''
    
    # get points
    n = imgs.shape[1]
    points = (Z1, np.arange(n), np.arange(n))
    
    # get interpolated points
    xi = np.rollaxis(np.mgrid[:n, :n], 0, 3).reshape((n**2, 2))
    xi = np.c_[np.repeat(Z1_interp,n**2), xi]
    #print(xi.shape)
    
    # interpolate
    img_interp = interpn(points, imgs, xi, method='linear', bounds_error=False, fill_value=None)
    img_interp = img_interp.reshape((n, n))
    #print(np.sum(img_interp))
    
    return(img_interp)


from scipy.interpolate import RegularGridInterpolator

def create_interpolator(images, Z1):
    if images.shape[0] != len(Z1):
        raise ValueError("Length of Z1 must match the first dimension of images.")
    
    # Define the interpolator
    interpolator = RegularGridInterpolator(
        points=(Z1, np.arange(images.shape[1]), np.arange(images.shape[2])),
        values=images,
        method='linear',
        bounds_error=False,
        fill_value=None
    )
    return interpolator


def interpolate_at(interpolator, Nimage, Z1_interp):
    """
    Interpolates the stack of images at the specified Z1_interp positions.

    INPUT:
    - Z1_interp: Single value or array of positions along Z1 to interpolate.

    OUTPUT:
    - Interpolated images at each Z1_interp position, with shape (len(Z1_interp), N, N)
      if Z1_interp is an array, or (N, N) if Z1_interp is a scalar.
    """
    Z1_interp = np.atleast_1d(Z1_interp)  # Ensure Z1_interp is array-like
    # Generate the grid of (i, j) positions for each Z1_interp
    grid = np.stack(np.meshgrid(np.arange(Nimage), np.arange(Nimage), indexing="ij"), axis=-1)
    grid = grid.reshape(-1, 2)  # Flatten (i, j) grid

    # Add the Z1_interp dimension to the grid
    full_grid = np.array([
        np.hstack((np.full((grid.shape[0], 1), z), grid))
        for z in Z1_interp
    ]).reshape(-1, 3)

    # Interpolate at all points
    interpolated = interpolator(full_grid).reshape(len(Z1_interp), Nimage, Nimage)
    return interpolated if len(Z1_interp) > 1 else interpolated[0]

    return interpolate_at
```
