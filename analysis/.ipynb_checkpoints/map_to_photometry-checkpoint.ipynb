## Coadd maps

```python
## function to add two maps

def add_maps(maps, headers, size_common):
    """
    need to pass wcs_are_consistent() check
    """
    
    # check which map is larger, truncate the larger one and add that onto the smaller one.
    if headers[0]['NAXIS1'] != headers[1]['NAXIS1']:
        bigger_idx = np.argmax([headers[0]['NAXIS1'], headers[1]['NAXIS1']])
        smallr_idx = np.argmin([headers[0]['NAXIS1'], headers[1]['NAXIS1']])
    else: # two maps are of the same size
        # arbiturally set indices...
        bigger_idx = 0
        smallr_idx = 1
        
    size_common = min(maps[smallr_idx].shape[0], size_common)
    
    # print(f"add map {smallr_idx} onto map {bigger_idx}...")
        
    # set the common central reference point based on the larger map
    x0, y0 = (headers[bigger_idx]['CRPIX1'], headers[bigger_idx]['CRPIX2']) 
    
    # calculate number of pixel span for the smaller map on the bigger map
    dx = headers[smallr_idx]['NAXIS1']//2 
    dy = headers[smallr_idx]['NAXIS2']//2
    
    # add the smaller map onto the larger one
    ymin = int(y0-dy)
    ymax = int(y0+dy)
    xmin = int(x0-dx)
    xmax = int(x0+dx)
    maps[smallr_idx] += maps[bigger_idx][ymin:ymax, xmin:xmax]
    
    return maps[smallr_idx], headers[smallr_idx], size_common
    


```

```python
### check all maps ref point and resolution
map_list = ['../../maps/schen6/deep_north_galaxies.fits', 
            '../../maps/schen6/deep_north_zodi.fits',
            '../../maps/schen6/deep_north_dc.fits',
            '../../maps/schen6/deep_north_read_noise.fits']

data_list = []
header_list = []
wcs_list = []

for file_path in map_list:
    with fits.open(file_path) as hdul:
        data = hdul[0].data
        header = hdul[0].header
        wcs = WCS(header)
        data_list.append(data)
        header_list.append(header)
        wcs_list.append(wcs)
        
```

```python
### sum up maps

def wcs_are_consistent(wcs1, wcs2):
    # only allow Npix to vary (also reference pixel)
    
    return (
        np.allclose(wcs1.wcs.crval, wcs2.wcs.crval) and  # ref coords
        np.allclose(wcs1.wcs.cdelt, wcs2.wcs.cdelt) and  # pixel resolution
        sorted(wcs1.wcs.ctype) == sorted(wcs2.wcs.ctype) # coordinate type
    )


assert all(wcs_are_consistent(wcs_list[0], wcs) for wcs in wcs_list), "WCS grids differ, reprojection needed."



for m in range(len(data_list)):
    
    if m==0:
        continue
        
    elif m==1:
        (map_summed, header_summed, size) = add_maps([data_list[m-1], data_list[m]], 
                                                     [header_list[m-1], header_list[m]],
                                                     size_common=100000) # give it a large value to start
    else:
        (map_summed, header_summed, size) = add_maps([map_summed, data_list[m]],
                                                     [header_summed, header_list[m]], 
                                                     size_common=size)
        
print(map_summed.shape, size)
print(header_summed)
```
